```python
import matplotlib.pyplot as plt

monthly_sales = df.groupby('Category')['Order_ID'].count()

plt.figure(figsize=(4, 4))
plt.pie(monthly_sales, labels=monthly_sales.index, autopct='%1.1f%%', startangle=140, 
        wedgeprops=dict(width=0.4, edgecolor='w'), pctdistance=0.85)

plt.title('Category Wise % of Total Sale', fontsize=14, fontweight='bold', pad=25)
plt.gca().set_aspect('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt

top_5_categories = df.groupby('Product_Name')['Order_ID'].count().nlargest(5).reset_index()

plt.style.use('fivethirtyeight')
plt.rcParams['font.family'] = 'Calibri'

plt.figure(figsize=(3, 5))
wedges, texts, autotexts = plt.pie(
    top_5_categories['Order_ID'], 
    labels=top_5_categories['Product_Name'], 
    autopct='%1.1f%%', 
    startangle=140, 
    wedgeprops=dict(width=0.4, edgecolor='w'),
    pctdistance=0.85
)
plt.gca().add_artist(plt.Circle((0, 0), 0.50, fc='white'))

for autotext in autotexts:
    autotext.set_fontsize(14)
    autotext.set_fontweight('bold')
for text in texts:
    text.set_fontsize(14)

plt.title('Top 5 Products Name by Order Count', fontsize=16, fontweight='bold')
plt.gca().set_aspect('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group by month and sum 'Order_Id'
monthly_data = df.groupby(df['Order_Date'].dt.to_period('M'))['Order_ID'].count()

plt.style.use('fivethirtyeight')
plt.rcParams['font.family'] = 'Calibri'

plt.figure(figsize=(15, 5))
plt.plot(monthly_data.index.astype(str), monthly_data, 
         marker='o', linestyle='-', color='b', 
         markersize=8, linewidth=2)

plt.title('Monthly Sales Trend', fontsize=18, fontweight='bold')
plt.xlabel('Month', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.grid(True, linestyle='--', linewidth=0.7)
plt.tight_layout()

plt.show()
```
